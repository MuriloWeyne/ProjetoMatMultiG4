```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def curva6(t):
    x = (21/2)*np.cos(t) - (21/4)*np.cos(6*t)
    y = (21/2)*np.sin(t) - (21/4)*np.sin(6*t)
    return x, y
```

```python
t = np.linspace(0, 2*np.pi, int(1e7), endpoint=False)
x, y = curva6(t)

plt.axis('equal')
plt.plot(x,y)
plt.grid()
plt.savefig("curva", dpi=180)
plt.show()
```

```python
u_v = [0.383, 0.874, 1.64, 2.13, 2.896, 3.387, 4.153, 4.643, 5.409, 5.9]

t = np.concatenate(
    [np.arange(u_v[i], u_v[i+1], 1e-3) for i in range(0, 10, 2)]
)
x, y = curva6(t)
plt.plot(x,y)

plt.axis('equal')
plt.grid()
plt.show()
```
